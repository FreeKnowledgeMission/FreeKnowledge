# Concate merge Dataframe

```python
import pandas as pd
```

```python
raw_data = {
    'member_id':['1','2','3','4','5'],
    'first_name':['Adam','Adrian','Alan','Alexander','Andrew'],
    'last_name':['Abraham', 'Allan','Alsop','Anderson','Arnold']}
df1 = pd.DataFrame(raw_data, columns = ['member_id','first_name','last_name'])
df1
```

```python
raw_data = {
    'member_id' :['4', '5', '6','7','8'],
    'first_name':['Benjamin','Blake','Boris','Brandon','Brian'],
    'last_name':['Baker','Ball','Bell','Berry','Black']}
df2 = pd.DataFrame(raw_data, columns = ['member_id','first_name','last_name'])
df2
```

# Concate two dataframes

```python
#row
pd.concat([df1,df2], axis=0)
```

```python
#columns
pd.concat([df1,df2], axis=1)
```

# Merge DataFrames

```python
#inner
pd.merge(df1,df2, on='member_id', how='inner')
```

```python
#outer
pd.merge(df1,df2, on='member_id', how='outer')
```

```python
#left
pd.merge(df1,df2, on='member_id', how='left')
```

```python
#right
pd.merge(df1,df2, on='member_id', how='right')
```
